```python
import numpy as np
import pandas as pd
```

```python
train_data=pd.read_csv("titanic/train.csv")
test_data=pd.read_csv("titanic/test.csv")
```

```python
train_data['Age']=train_data['Age'].fillna(train_data['Age'].mean())
train_data['Embarked']=train_data['Embarked'].fillna(train_data['Embarked'].mode())
train_data=train_data.drop(['Cabin'],axis=1)
train_data.isna().sum()
```

```python
test_data['Age']=test_data['Age'].fillna(test_data['Age'].mean())
test_data['Fare']=test_data['Fare'].fillna(test_data['Fare'].mean())
test_data=test_data.drop(['Cabin'],axis=1)
test_data.isna().sum()
```

```python
import matplotlib.pyplot as plt
train_data.boxplot()
plt.show()
```

```python
for i in ['Age','Fare']:
 q1=train_data[i].quantile(0.25)
 q3=train_data[i].quantile(0.75)
#print(q1,q3)
 iqr=q3-q1
 lower_limit = q1 - 1.5*iqr
 upper_limit = q3 + 1.5*iqr
#print(lower_limit,upper_limit)
 train_data[i] = np.where(train_data[i]>upper_limit,upper_limit,train_data[i])
 train_data[i] = np.where(train_data[i]<lower_limit,lower_limit,train_data[i])
```

```python
import matplotlib.pyplot as plt
train_data.boxplot()
plt.show()
```

```python
train_data.info()
```

```python
def min_max_scaling(column):
    return ((column-column.min())/(column.max()-column.min())*(1-0)+0)

numerical_cols=['Age','Fare']

for i in numerical_cols:
    train_data[i]=min_max_scaling(train_data[i])
```

```python
train_data = pd.get_dummies(train_data,columns=['Sex'])
train_data.head()
```

```python
test_data = pd.get_dummies(test_data,columns=['Sex'])
test_data.head()
```

```python
x_train=train_data.iloc[:,[2,4,8,11]]
y_train=train_data.iloc[:,[1]]
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LogisticRegression  
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
```

```python
test_data.info()
```

```python
x_test=test_data.iloc[:,[1,3,7,10]]
```

```python
y_pred=classifier.predict(x_test)
```
